Ada opini bahwa harga mobil dipengaruhi oleh warna dari mobil. Apakah pernyataan tersebut benar?

```python
df1=df.drop(columns=['Nickname','Car Gender','Buyer Gender','Buyer Age','Buzzword','Country','City','Dealer Latitude','Dealer Longitude','New Car','Purchase Date','Discount','5-yr Depreciation','Top Speed'])
df1.sort_values("Sale Price",ascending=False).head(5)
```

```python
def color_max(s):
    is_max = s == s.max()
    return ['background-color: lightblue' if v else '' for v in    is_max]
```

```python
df1.groupby(["Color"]).sum().sort_values("Sale Price",ascending=False).head(5)
mean = df1.groupby(["Color"]).mean().sort_values("Sale Price",ascending=False).head(5)
mean.style.apply(color_max, subset=['Sale Price','Resell Price'])
```

Dapat dilihat jika berdasarkan harga per unit Blue, Orange, Pink,serta maroon merupakan warna yang memiliki Sale Price paling tinggi, namun jika dilihat berdasarkan rata ratanya Orange, Pink,serta maroon tidak berada di posisis 5 teratas.
Ini berarti Harga dari mobil berwarna Orange,Pink,serta maroon tidak rata, yang menunjukkan bahwa warna mobil tidak mempengeruhi harga beli mobil.

Berdasarkan Persentasi dapat kita lihat penjualan mobil dengan warna Merah , Khaki, Goldenrod, Fuscia, serta Indigo merupakan warna mobil yang dijual kembali yang mengalami penurunan paling rendah

Mobil dengan warna Goldenrod mengalami penurunan harga sebesar 30.5% saat dijual kembali. Angka ini lebih rendah dibanding angka rata-rata persetasi penurunan harga jual mobil. Bahkan ada mobil dengan warna Goldenrod dijual kembali dengan harga hampir 5x harga belinya. Untuk itu kami menyarankan untuk membeli mobil dengan warna Goldenrod karena warna itu memiliki nilai yang lebih baik ketika dijual, meskipun ketika dijual dengan kondisi telah dipakai beberapa tahun.
